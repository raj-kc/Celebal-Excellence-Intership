## Step 3: Data Cleaning <a id='step3'></a>
Tasks performed:

- Convert price columns into numeric format
- Handle missing values
- Remove duplicate records

```python
# 3a. Convert price-related columns to numeric format
# The 'final_price' column contains strings like: "₹3,995.00"
# We need to:
#   1. Strip surrounding quotes
#   2. Remove the ₹ currency symbol
#   3. Remove comma separators
#   4. Convert to float

# Sample before cleaning
print("Before cleaning:")
print(df['final_price'].head())

# Clean the final_price column step by step
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('"', '', regex=False)   # Remove surrounding quotes
    .str.replace('₹', '', regex=False)   # Remove Rupee symbol
    .str.replace(',', '', regex=False)   # Remove commas (e.g., 3,995 → 3995)
    .str.strip()                         # Remove whitespace
)

# Convert to numeric — 'coerce' turns any invalid values into NaN
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

print("\nAfter cleaning:")
print(df['final_price'].head())
print(f"\n final_price dtype: {df['final_price'].dtype}")
```

```python
# 3b. Handle Missing Values
# Strategy per column:
#   - 'discount'           → Fill with 0 (no discount if not mentioned)
#   - 'what_customers_said' → fill with 'No Feedback Provided'
#   - 'seller_name'        → Fill with 'Unknown'
#   - 'seller_information' → Fill with 'Unknown'
#   - 'videos'             → Fill with 'None' (product has no video)
#   - 'variations'         → Fill with 'None'

# Fill numeric missing values
df['discount'] = df['discount'].fillna(0)

# Fill categorical missing values with placeholders
df['seller_name']         = df['seller_name'].fillna('Unknown')
df['seller_information']  = df['seller_information'].fillna('Unknown')
df['videos']              = df['videos'].fillna('None')
df['variations']          = df['variations'].fillna('None')
df['what_customers_said'] = df['what_customers_said'].fillna('No Feedback Provided')

# Confirm result
remaining_nulls = df.isnull().sum().sum()
print(f"Missing values handled. Remaining nulls in dataset: {remaining_nulls}")
```

```python
# 3c. Remove Duplicate Records
# Duplicates can skew aggregations and analysis.
# We check based on 'product_id' since each product should be unique.

# 1. Count the number of duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")

# 2. If duplicates are found, remove them
if duplicate_count > 0:
    # Remove duplicate rows and keep the first occurrence
    df.drop_duplicates(keep='first', inplace=True)
    print("Success: All duplicate rows have been removed!")
    print(f"New dataset shape: {df.shape}")
else:
    print("No duplicates found. Your dataset is already clean!")
```

## Step 4: Feature Engineering <a id='step4'></a>

Creating new derived columns that add analytical value to the dataset.

```python
# 4a. Price Difference Column
#     = initial_price - final_price
#     Shows the actual monetary savings per product

df['price_difference'] = df['initial_price'] - df['final_price']

print("'price_difference' column created.")
df[['product_id', 'initial_price', 'final_price', 'price_difference']].head()
```

```python
# 4b. Popularity Metric
#     Formula: popularity_score = rating × log1p(ratings_count)
#     Why log1p?
#     A product with 5 stars but only 2 ratings shouldn't outrank
#     a product with 4.5 stars and 4000 ratings.
#     log1p(ratings_count) compresses large counts to avoid domination.

df['popularity_score'] = df['rating'] * np.log1p(df['ratings_count'])

# Round to 2 decimal places for readability
df['popularity_score'] = df['popularity_score'].round(2)

print("'popularity_score' column created.")
print(f"   Score range: {df['popularity_score'].min():.2f} — {df['popularity_score'].max():.2f}")
print()

# Top 5 most popular products
print("Top 5 Most Popular Products:")
df[['product_id', 'title', 'rating', 'ratings_count', 'popularity_score']]\
  .sort_values('popularity_score', ascending=False).head(5)
```

```python
# 4c. Price Tier Column (derived from final_price)
#     Buckets products into Budget / Mid-Range / Premium / Luxury
#     Useful for category-level and segment-level analysis

def price_tier(price):
    """Classify product price into a tier."""
    if price < 500:
        return 'Budget'
    elif price < 1500:
        return 'Mid-Range'
    elif price < 5000:
        return 'Premium'
    else:
        return 'Luxury'

df['price_tier'] = df['final_price'].apply(price_tier)

print("'price_tier' column created.")
print()
print("Distribution of Price Tiers:")
print(df['price_tier'].value_counts())
```

## Step 5b: Bivariate Analysis <a id='step5b'></a>

Exploring relationships between pairs of variables.



```python
# Rating vs Final Price — Does higher price mean better ratings?

plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x='final_price',
    y='rating',
    scatter_kws={'alpha':0.4},
    line_kws={'color':'red'}
)

plt.title('Relationship Between Final Price and Rating')
plt.xlabel('Final Price')
plt.ylabel('Rating')

plt.show()

# Correlation coefficient

correlation = df['final_price'].corr(df['rating'])

print(f"Correlation between Final Price and Rating: {correlation:.2f}")
```

The correlation coefficient (0.08) indicates a negligible positive relationship between Final Price and Rating, suggesting that expensive products are not necessarily rated higher by customers.
    

```python
# Discount % vs Price Difference (actual savings in ₹)

plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=df, x='discount', y='price_difference',
    alpha=0.4, color='coral', edgecolor='white'
)
plt.title('Discount (%) vs Actual Price Savings (₹)', fontsize=14)
plt.xlabel('Discount (%)')
plt.ylabel('Price Difference (₹)')
plt.savefig('plot_discount_vs_savings.png', dpi=150, bbox_inches='tight')
plt.show()

correlation = df['discount'].corr(
    df['price_difference']
)

print(f"Correlation between Discount % and Price Difference: {correlation:.2f}")
```

```python
# Correlation Heatmap — visualize relationships across all numeric columns

numeric_cols = ['rating', 'ratings_count', 'initial_price',
                'discount', 'final_price', 'price_difference', 'popularity_score']

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(
    corr_matrix, annot=True, fmt='.2f',
    cmap='coolwarm', center=0, linewidths=0.5
)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.tight_layout()
plt.savefig('plot_correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Step 5c: Category-Level Analysis <a id='step5c'></a>

Aggregating data by product category to uncover segment-level patterns.

```python
# ─────────────────────────────────────────────────────────────────────────────
# Aggregate metrics by category
# ─────────────────────────────────────────────────────────────────────────────

category_stats = df.groupby('category').agg(
    product_count        = ('product_id', 'count'),
    avg_rating           = ('rating', 'mean'),
    avg_final_price      = ('final_price', 'mean'),
    avg_discount_pct     = ('discount', 'mean'),
    avg_popularity_score = ('popularity_score', 'mean'),
    avg_price_diff       = ('price_difference', 'mean')
).round(2).sort_values('product_count', ascending=False)

print("Category-Level Summary (Top 15 by Product Count):")
category_stats.head(15)
```

```python
# Top 5 Categories by Average Rating

top_rated = category_stats.sort_values('avg_rating', ascending=False).head(5)
top_rated
```

```python
# Top 10 Most Expensive Categories (by avg final price)
most_expensive = category_stats.sort_values('avg_final_price', ascending=False).head(10)
most_expensive.head()
```

## Step 6: Visualizations <a id='step6'></a>

```python
# Price Tier Distribution (Pie Chart)

tier_counts = df['price_tier'].value_counts()
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

plt.figure(figsize=(7, 7))
plt.pie(
    tier_counts.values,
    labels=tier_counts.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)
plt.title('Distribution of Products by Price Tier', fontsize=14)
plt.savefig('plot_price_tier_pie.png', dpi=150, bbox_inches='tight')
plt.show()
```

### Observation

Most products fall under the **Mid-Range** category (47.3%), followed by **Premium** (33.6%). Luxury products make up only a small portion (4.0%) of the dataset.

### Insight

The dataset is dominated by mid-range and premium products, suggesting that these segments are the primary focus of the product catalog.

### Observation

Ratings are generally similar across Budget, Mid-Range, Premium, and Luxury products, with median ratings around 4.0.

### Insight

Higher-priced products do not receive significantly better ratings, indicating that price is not a major driver of customer satisfaction.
